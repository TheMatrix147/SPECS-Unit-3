#### Fill in blanks with a brief description of each plot type:

  
- Histogram
> Histograms show frequency distributions. They are best used for displaying continuous data.
- Density Plot / KDE
> A KDE plot displays the probability density function of a continuous dataset. It is best used for visualizing the data's probability density. 
- Bar Chart
> A bar chart has two axes: on the x axis is the category, and on the y axis is the category's value. The bar chart is best used for displaying segments of categorical variable information.
- Line Plot
> In a line plot, the distribution of data is represented by data points that are connected by line segments. Line plots are best for analyzing changes in data over a period of time. 
- Univariate scatter plot
> A Univariate scatter plot shows a single variable's values. These values, represented by dots, are plotted according to their observation number. These plots are best used to provide insight on a single data variable. 
- Strip Plot
> A strip plot displays a scatterplot where one variable is categorical. It is best used for creating subsets that correspond to the categorical variable.
- Violin Plot
> Much like a box plot, a violin plot also shows a 5-number summary, but it also contains a rotated kernal density plot. Violin plots are best used to analyze the distribution of numeric data. 
- Box Plot
> A box plot shows a 5-number summary of a dataset, consisting of the minimum, first quartile, median, third quartile, and maximum. A box plot is best used for showing a data set's distribution given this summary. 
- Pie Chart
> A pie chart displays a distribution of data in a circular format, with different slices of the circle representing different categories. A pie chart is best used to observe numerical sample data. 
- Rug Plot
> A rug plot shows a distribution of data based off of tick marks that are plotted on just one axis. These tick marks represent the data points. Rug plots are best used for single quantitative variables.

### Step 3 - Use Pearsonr & Spearmanr To Answer Questions About Correlation and P-Values

```python
# load salary_numeric.csv into dataframe where education levels have been re-expressed as numeric
```

```python
################################### TO DO #########################################
####### FOLLOW PROJECT INSTRUCTIONS TO COMPLETE THIS STEP IN THE CELLS BELOW ######
```

```python
# import pearson and spearmanr methods from scipy.stats library
```

```python
from scipy.stats import pearsonr, spearmanr
```

### What is the correlation between salary and education?

#### Your Hypothesis = The higher the education, the higher the salary.

```python
# calculate correlation and pvalue of salary and education with spearmanr method
```

```python
spearmanr(df['salary'], df['education'])
```

There is a moderate positve correlation between education level and salary.

#### Are these results statistically significant? How do you know?

Yes. They are highly statistically significant because the p value is less than 0.01. 

### What is the correlation between salary and age?

#### Your Hypothesis = The higher the age, the higher the salary. 

```python
# calculate correlation and pvalue of salary and age with pearsonr method
```

```python
pearsonr(df['salary'], df['age'])
```

There is a strong positive correlation between salary and age.

#### Are these results statistically significant? How do you know? 

Yes. They are statistically significant because the p value is less than 0.05.

### What is the correlation between salary and industry experience? 

#### Your Hypothesis = The more industry experience one has, the higher that person's salary. 

```python
# calculate correlation and pvalue of salary and industry experience with pearsonr method
```

```python
pearsonr(df['salary'], df['industry_experience'])
```

There is a strong positive correlation between salary and industry experience.

#### Are these results statistically significant? How do you know?

Yes. They are statistically significant because the p value is less than 0.05.

Yes. All three results showed positive correlations that were statistically significant. Thus, the null hypothesis can be rejected in all three cases.
